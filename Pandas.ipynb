## Map and ApplyMap


### assign() - 👉 Use when you want to add/transform columns.

```python
import pandas as pd

df = pd.DataFrame({
    "name": ["Asim", "Ali", "Mohd"],
    "age": [25, 30, 22]
})

# Add a new column "age_plus_5"
df2 = df.assign(age_plus_5 = df["age"] + 5)

print(df2)
```

### map() - 👉 Use when you want to transform one column (Series).

For a selected column , it goes to every cells of that column
Applies a function, dictionary, or mapping to each value in the column.



```python
# Map using a function
df["age_squared"] = df["age"].map(lambda x: x**2)

# Map using a dictionary
df["short_name"] = df["name"].map({"Asim": "AS", "Ali": "AL", "Sara": "SR"})

print(df)
```

### applymap() - Apply function to every cell of DataFrame.
Takes a function and applies it to each value, not row/column.

```python
df_num = pd.DataFrame({
    "A": [1, 2, 3],
    "B": [4, 5, 6]
})

# Applymap to square every value
df_squared = df_num.applymap(lambda x: x**2)

print(df_squared)
```

# Data Grouping and Aggregation

## Grouping and then apply aggregate function

```python
import pandas as pd

data = {
    "OrderID": [1, 2, 3, 4, 5, 6, 7, 8],
    "Product": ["Shirt", "Shirt", "Pants", "Pants", "Shoes", "Shoes", "Shirt", "Pants"],
    "Category": ["Clothing", "Clothing", "Clothing", "Clothing", "Footwear", "Footwear", "Clothing", "Clothing"],
    "Quantity": [2, 1, 3, 2, 1, 4, 5, 1],
    "Price": [500, 500, 800, 800, 2000, 2000, 500, 800],
    "Sales": [1000, 500, 2400, 1600, 2000, 8000, 2500, 800]
}

df = pd.DataFrame(data)
print(df)
```

```python
result = df.groupby("Category").agg(
    Total_Sales=("Sales", "sum"),
    Avg_Sales=("Sales", "mean"),
    Total_Quantity=("Quantity", "sum"),
    Num_Orders=("OrderID", "count"),
    Num_Products=("Product", "nunique")   # Unique products per category
).reset_index()
```

```python
print(result)
```

## Join

### Inner Join

```python
result = pd.merge(orders, customers, on="CustomerID", how="inner")
print(result)
```

### LEFT JOIN → all from orders, match from customers

```python
result = pd.merge(orders, customers, on="CustomerID", how="left")
print(result)
```

### RIGHT JOIN → all from customers, match from orders

```python
result = pd.merge(orders, customers, on="CustomerID", how="right")
print(result)
```

### OUTER JOIN → all records from both sides

```python
result = pd.merge(orders, customers, on="CustomerID", how="outer")
print(result)
```

```python
import pandas as pd
import numpy as np

data = {
    "OrderID": [101, 102, 103, 104, 105, 106],
    "Customer": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank"],
    "Product": ["Shoes", "Shirt", "Shoes", "Pants", "Shoes", "Shirt"],
    "Quantity": [2, np.nan, 1, 3, np.nan, 2],
    "Price": [50, 20, 50, np.nan, 50, 20],
    "Sales": [100, np.nan, 50, 90, np.nan, 40]
}

df = pd.DataFrame(data)
print(df)
```

### 1. Fill with constant values - Useful for categorical fields or fixed replacements.

```python
df["Quantity"].fillna(0, inplace=True)
df["Sales"].fillna(0, inplace=True)
```

### 2. Forward Fill / Backward Fill
Use previous or next row value (good for time-series).

```python
df["Price"] = df["Price"].ffill()   # forward fill
df["Price"] = df["Price"].bfill()   # backward fill
```

### 3. Fill with mean/median/mode (numeric data)
Best for continuous values.

```python
df.fillna({
    "Quantity": df["Quantity"].mean(),
    "Price": df["Price"].median()
}, inplace=True)
```
