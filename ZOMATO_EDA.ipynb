```python
country_value = final_df.Country.value_counts().values
country_value
```

## Pie Chart

## CHANGING COLUMN NAME 0 to Rating Count IN DATA FRAME (.rename())

```python
df_ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
df_ratings
```

## VISUALISING THE RATINGS
OBSERVATION:
1. WHEN THE RATING IS BETWEEN 4.5 TO 4.9 -- EXCELLENT
2. WHEN THE RATING IS BETWEEN 4.0 TO 3.4 -- VERY GOOD
3. WHEN THE RATING IS BETWEEN 3.5 TO 3.9 -- GOOD
4. WHEN THE RATING IS BETWEEN 3.0 TO 3.4 -- AVERAGE
5. WHEN THE RATING IS BETWEEN 2.5 TO 3.9 -- AVERAGE
6. WHEN THE RATING IS BETWEEN 2.0 TO 2.4 -- POOR

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y = "Rating Count" , data = df_ratings)
```

# BAR WITH DIFFERENT COLOR


```python
sns.barplot(x="Aggregate rating",y = "Rating Count" ,hue ='Rating color', data = df_ratings)
```

# Mapping THE RATING COLOR WITH THE BAR PLOT


```python
#Selecting One Column in DataFrame and Getting Distinct Values:
distinctRatingColor = df_ratings['Rating color'].unique().tolist()
distinctratingColor
```

```python
palette = []
for i in range(0,len(distinctRatingColor)):
    palette.append(distinctRatingColor[i])

# Replace all occurrences of "DarkGreen" with "Green"
colors = ["Green" if item == "Dark Green" else item for item in palette]
colors
```

```python
sns.barplot(x="Aggregate rating",y = "Rating Count" ,hue = 'Rating color', data = df_ratings , palette = colors)
```

OBSERVATION:
1. Not rated count is very high
2. Maximum number of Rating are between 2.5 to 3.4
3. As maximum rating lies between 2.5 to 3.4 so we can replace missing rating values with values present between it.
   

## Count Plot

```python
sns.countplot(x='Rating color', hue='Rating color', data=df_ratings, palette=colors, legend=False)
```
