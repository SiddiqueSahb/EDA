
**Numpy is faster than list**

```python
import time
import numpy as np
size = 100000
def addition_using_list():
  t1 = time.time()
  a = range(size)
  b = range(size)
  c = [a[i] + b[i] for i in range(size)]
  t2 = time.time()
  return t2 - t1

def addition_using_numpy():
  t1 = time.time()
  a = np.arange(size)
  b = np.arange(size)
  c = a + b
  t2 = time.time()
  return t2 - t1
t_list = addition_using_list()
t_numpy = addition_using_numpy()
print("List = ", t_list * 1000)
print("NumPy = ", t_numpy * 1000)
```

# **np.linspace()**

numpy.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None, axis=0)[source]

>
This returns evenly spaced numbers over a specified interval.

The endpoint of the interval can optionally be excluded.

Parameters

**start** : array_like The starting value of the sequence.

**stop** : array_like

The end value of the sequence, unless endpoint is set to False. In that case, the sequence consists of all but the last of num + 1 evenly spaced samples, so that stop is excluded. Note that the step size changes when endpoint is False.

**num** : int, optional

Number of samples to generate. Default is 50. Must be non-negative.

**endpoint** : bool, optional

If True, stop is the last sample. Otherwise, it is not included. Default is True.

**retstep** : bool, optional

If True, return (samples, step), where step is the spacing between samples.

**dtype** : dtype, optional

The type of the output array. If dtype is not given, the data type is inferred from start and stop. The inferred dtype will never be an integer; float is chosen even if the arguments would produce an array of integers.

**axis** : int, optional

The axis in the result to store the samples. Relevant only if start or stop are array-like. By default (0), the samples will be along a new axis inserted at the beginning. Use -1 to get an axis at the end.

```python
a = np.linspace(2,10)
print(a)
#it generates sample space by default = 50
```

```python
a = np.linspace(2,10,5,dtype = float)
print(a)
#prints start from 2 till 10 and having 5 samples
```

```python
a = np.linspace(2,10,5,dtype = int,endpoint = False )
print(a)
#endpoint false so dont include 10
#In linspace we have control over the number of values
```

# **np.sort()**

sort() Arguments

The sort() method takes four arguments:
>
**array** - array to be sorted
 >

**Sort a Multidimensional Array**
>
Multidimensional arrays are sorted based on the given axis.

```python
import numpy as np

array = np.array([[3, 10, 2], [1, 5, 7], [2, 7, 5]])
print("Matrix before sorting")
print(array)
# sort column wise based on the axis 1
array2 = np.sort(array)

# flattens the given array and sorts the flattened array
array3 = np.sort(array, axis = None)

# sort array row wise
array4 = np.sort(array, axis = 0) #each row 1st element is sorted first (i.e first column is sorted)

print('Sorted based on last axis(1): \n', array2)
print('Sorted a flattened array', array3)
print('Sorted based on axis 0: \n', array4)
```

# **Numpy Broadcasting**

**Suppose there is 2 - 2D matrix of same dimension then we can perform mathematical operations directly matrix3 = matrix1 +(operator) matrix2
**
>
Broadcasting - means that how numpy will treat matrix of different(unequal) dimensions
We can't simply perform mathematical operation . Element by element mathematical operation is not possible.




```python
import numpy as np

x = np.random.randint(1,10,(3,3)) #ARRAY 1
y = np.random.randint(1,10,(3,3)) #ARRAY 2
print("MATRIX 1")
print(x)
print("MATRIX 2")
print(y)
print("ANS AFTER ADDITION")
z = x+y
print(z)
```

**ADDITION OF MATRIX OF UNEQUAL DIMENSION**

```python
x = np.random.randint(1,10,(3,3)) #ARRAY 1 (3 X 3)
y = np.random.randint(1,10,(3)) #ARRAY 2  (1 X 3)
# if matrix is (3 x 1) the repeat column 2 times
print("MATRIX 1")
print(x)
print("MATRIX 2")
print(y)
#DUE TO BROADCASTING MATRIX 2 WILL BE REPEATED 2 TIMES WHICH WILL MAKE ITS DIMENSION
#EQUAL TO MATRIX 2
"""
#BROADCASTING WILL DO THIS THEN ON THIS ARRAY MATHEMATICAL OPERATION IS PERFORMED
6 7 8
6 7 8
6 7 8
"""
print("ANS AFTER ADDITION") #DIMENSION OF ANS MATRIX (3 X 3 ) X (1 X 3) = (3 X 3)
z = x+y
print(z)
```

**EX-2**

ERROR DUE TO UNEQUAL SIZE.
SOLUTION- SOLVING BY DOING TRANSPOSE OF MATRIX

```python
x = np.random.randint(1,10,(3,2)) #ARRAY 1
y = np.random.randint(1,10,(2,3)) #ARRAY 2
print("MATRIX 1")
print(x)
print("MATRIX 2")
print(y)
print("ANS AFTER ADDITION")
z = x+y
print(z)
```

**TRANSPOSE**

```python
x = np.random.randint(1,10,(3,2)) #ARRAY 1
y = np.random.randint(1,10,(2,3)) #ARRAY 2
print("MATRIX 1")
print(x)
print("MATRIX 2")
print(y)
print("MATRIX AFTER TRANSPOSE ")
y = np.transpose(y)
print(y)
print("ANS AFTER ADDITION")
z = x+y
print(z)
```
